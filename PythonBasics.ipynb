```python
 #6 Write a function to demonstrate the behavior of mutable and immutable arguments.
def mutability(mutable_list, immutable_int):
  print(f"mutable: {mutable_list}, immutable: {immutable_int}")
  mutable_list.append("modified")
  immutable_int += 100
  print(f"changes - mutable: {mutable_list}, immutable: {immutable_int}")

my_list = [1,3,6]
my_int = 80
# print(f"mutable: {my_list}, immutable: {my_int}")
mutability(my_list, my_int)
print(f"mutable: {my_list}, immutable: {my_int}")
```

```python
#10 Write code to demonstrate type casting with list elements.
mixed_list = [10, 20, 30.8, "10"]

print("Original list:", mixed_list)
print("Original types:", [type(x) for x in mixed_list])

# Cast all to integers using map()
str_list = list(map(str, mixed_list))
print("String cast:", str_list)
print("New types:", [type(x) for x in str_list])
```

```python
#14  Write a program to reverse a string using a while loop.
String = "Aryan"
rev = ""
idx = len(String)-1
while idx >= 0:
  rev += String[idx]
  idx-=1
print(rev)
```

```python
#15 Write a Python program to calculate the factorial of a number provided by the user using a while loop
i = int(input("Enter number "))
fact = 1
while i > 0:
  fact = fact * i
  i=i-1
print(fact)
```
