Q1. What is NumPy, and why is it widely used in Python?
 -NumPy (Numerical Python) is the foundational package for scientific computing in Python, providing support for large, multi-dimensional arrays and matrices.

Q2. How does broadcasting work in NumPy?
 -NumPy broadcasting is a mechanism that allows arithmetic operations on arrays of different shapes and sizes by implicitly expanding the smaller array to match the larger one's shape.

Q3. What is a Pandas DataFrame?
 -A Pandas DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure in Python.

Q4. Explain the use of the groupby() method in Pandas.
 -It allows to organize data into groups based on one or more criteria (e.g., column values) and then perform a range of operations on each group independently.

Q5.  Why is Seaborn preferred for statistical visualizations?
 -Seaborn is preferred for statistical visualization because it simplifies creating complex, publication-quality graphics with minimal code, offering a high-level interface over Matplotlib.

Q6. What are the differences between NumPy arrays and Python lists?
 -NumPy arrays are highly optimized for numerical operations and large datasets, while Python lists offer more flexibility in data types. Numpy arrays are faster for numerical and vector operations.

Q7. What is a heatmap, and when should it be used
 -A heatmap is a data visualization technique that represents magnitudes of phenomena as colors in two dimensions, allowing for the quick identification of trends, patterns, and anomalies within complex datasets.

Q8. What does the term “vectorized operation” mean in NumPy?
 -In NumPy, a vectorized operation refers to the ability to perform mathematical calculations on entire arrays of data at once.

Q9. How does Matplotlib differ from Plotly?
 -Matplotlib is a foundational, low-level library for creating highly customizable, static plots, ideal for academic publications, while Plotly is a newer, high-level library designed for creating interactive, web-ready visualizations, best suited for dashboards and web applications.

Q10. What is the significance of hierarchical indexing in Pandas?
 -Its primary significance lies in efficiently organizing data, simplifying complex queries, and facilitating sophisticated data analysis.

Q11. What is the role of Seaborn’s pairplot() function?
 -Seaborn's pairplot() function is a powerful tool for Exploratory Data Analysis (EDA) that creates a grid of plots to visualize the pairwise relationships between multiple variables in a dataset.

Q12. What is the purpose of the describe() function in Pandas?
 -The describe() function in the Python pandas library is used to generate descriptive statistics that summarize the central tendency, dispersion, and shape of a dataset's distribution, excluding NaN values.

Q13. Why is handling missing data important in Pandas?
 -Handling missing data in Pandas is crucial because it ensures the accuracy, integrity, and reliability of data analysis and subsequent machine learning models.

Q14. What are the benefits of using Plotly for data visualization?
 -It allows for zooming, panning, and hover-data exploration, transforming static charts into dynamic insights, which makes it ideal for data science, dashboards, and complex data analysis.

Q15. How does NumPy handle multidimensional arrays?
 -NumPy handles multidimensional arrays using its core object, the ndarray (N-dimensional array), which stores homogeneous data in a contiguous block of memory and uses metadata to interpret its shape and dimensions.

Q16. What is the role of Bokeh in data visualization?
 -Bokeh is an open-source Python library designed for creating interactive, web-ready visualizations, ranging from simple plots to complex, dynamic dashboards.

Q17. Explain the difference between apply() and map() in Pandas.
 -The primary difference is that map() operates exclusively on a pandas Series (a single column), while apply() can operate on a Series or an entire DataFrame (rows or columns at a time).

Q18. What are some advanced features of NumPy?
 -These features include sophisticated indexing techniques, specialized array types, and optimized operations.

Q19. How does Pandas simplify time series analysis?
 -Pandas simplifies time series analysis by providing specialized data structures and a robust suite of optimized tools for managing and manipulating time-stamped data.

Q20. What is the role of a pivot table in Pandas?
 -A pivot table in Pandas functions as a powerful data summarization tool that reshapes, rearranges, and aggregates large datasets to reveal hidden trends.

Q21.  Why is NumPy’s array slicing faster than Python’s list slicing?
 -NumPy array slicing is significantly faster than Python list slicing primarily because NumPy slicing creates a view of the original data rather than a new copy, and because NumPy arrays are stored in a contiguous block of memory.

Q22. What are some common use cases for Seaborn?
 -Common use cases revolve around exploratory data analysis (EDA), understanding data patterns, and communicating insights through visually appealing plots.

```python
#4 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load sample dataset (or use your own DataFrame)
df = sns.load_dataset('iris').select_dtypes(include='number')  # Numeric columns only

# Calculate correlation matrix
corr_matrix = df.corr()

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.tight_layout()
plt.show()
```

```python
#8 Create a line plot with multiple lines using Matplotlib
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(0, 10, 100)

# Create multiple y datasets
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) + np.cos(x)

# Plot all lines
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)', linestyle='--')
plt.plot(x, y3, label='sin(x)+cos(x)', linewidth=2)

plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Multiple Line Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#10 Create a histogram using Seaborn to visualize a distribution

# Load sample data (or use your own DataFrame/array)
df = sns.load_dataset('iris')

# Create histogram
sns.histplot(data=df, x='sepal_length', bins=20, kde=True)
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
# Histogram by species (hue parameter)
sns.histplot(data=df, x='sepal_length', hue='species', bins=20,
             kde=True, element='step')  # 'step' avoids overlap
plt.title('Sepal Length by Species')
plt.show()

```

```python
#13 Create a 3D scatter plot using Plotly
import plotly.graph_objects as go

# Generate sample data
np.random.seed(42)
n_points = 100
x = np.random.normal(0, 1, n_points)
y = np.random.normal(0, 1, n_points)
z = x * y + np.random.normal(0, 0.1, n_points)

# Create 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x, y=y, z=z,
    mode='markers',
    marker=dict(
        size=5,
        color=z,  # Color by z values
        colorscale='Viridis',
        opacity=0.8
    ),
    hovertemplate='<b>X</b>: %{x:.2f}<br>' +
                  '<b>Y</b>: %{y:.2f}<br>' +
                  '<b>Z</b>: %{z:.2f}<extra></extra>'
)])

fig.update_layout(
    title='Interactive 3D Scatter Plot',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)
fig.show()
```
