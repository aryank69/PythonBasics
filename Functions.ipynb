Q5. What are iterators in Python and how do they differ from iterables?
 -An iterable is a Python object capable of returning its members one at a time, while an iterator is an object that represents a stream of data and can produce the next value in the sequence using the next() function.

Q6. Explain the concept of generators in Python and how they are defined.
 -Generators in Python are a simple tool used to create iterators, they produce items one by one and only when requested.

Q7. What are the advantages of using generators over regular functions?
 -Generators are superior to regular functions for large/infinite data because they use lazy evaluation, yielding values one by one instead of storing everything in memory, leading to huge memory savings.

Q8. What is a lambda function in Python and when is it typically used?
 -A lambda function in Python is a small, anonymous function defined without a name using the lambda keyword. It can take any number of arguments but can only have one expression.

Q9. Explain the purpose and usage of the `map()` function in Python.
 -It used to apply a specific function to every item in an iterable such as a list, tuple, etc.

Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
 -map():- Applies a function to every item in an iterable.
  filter():- Selects items from an iterable for which a function is true.
  reduce():- Applies a function cumulatively to the items of an iterable, reducing it to a single final value.

```python
#1 Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list
def even(nums):
  total = 0
  for i in nums:
    if i%2 == 0:
      total += nums
  return total

```

```python
#3  Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def sq(num):
  return list(map(lambda x:x**2, num))
```

```python
#4 Write a Python function that checks if a given number is prime or not from 1 to 200
def is_prime(pm):
  if pm<=1:
    return False
  for i in range(2, 200):
    if(pm % i) == 0:
      return False
  return True
```

```python
#5  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
def fib(n):
  a=0
  b=1
  for i in range(n):
    yield a
    a,b = b, a+b
```

```python
#6 Write a generator function in Python that yields the powers of 2 up to a given exponent
def expo(max_power):
  for i in range(max_power + 1):
    yield 2**i
```

```python
#7 Implement a generator function that reads a file line by line and yields each line as a string
def read_file(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except IOError as e:
        print(f"An I/O error occurred: {e}")
```

```python
#8  Use a lambda function in Python to sort a list of tuples based on the second element of each tuple
data = [('apple', 50, 30), ('banana', 25, 87), ('Cherry', 55, 12)]
data.sort(key=lambda x:x[1])
print(data)
```

```python
#9 Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit
def conv(celsius):
  return (celsius * 9/5) + 32
celsius_temperatures = [0, 10, 20, 30, 40, 100]
fahrenheit_temperatures = list(map(conv, celsius_temperatures))
print(fahrenheit_temperatures)
```

```python
#10 Create a Python program that uses `filter()` to remove all the vowels from a given string
def remove_vowels_filter(input_string):
    vowels = "aeiouAEIOU"

    def is_consonant(char):
        return char not in vowels
    filtered_characters = list(filter(is_consonant, input_string))
    result_string = "".join(filtered_characters)

    return result_string
```

```python
#11 Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
#order is smaller than 100,00 €.
orders = [
    [34587, 4, 40.95],
    [98762, 5, 56.50],
    [77226, 3, 32.95],
    [88112, 3, 24.99]
]
calculate_total = lambda order: (order[0], (order[1] * order[2]) + (10 if order[1] * order[2] < 100 else 0))

result = list(map(calculate_total, orders))

print(result)
```
